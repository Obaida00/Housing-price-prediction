```python
from os.path import exists
import pandas as pd

trainer = pd.read_csv("https://github.com/Obaida00/Housing-price-prediction/raw/main/train.csv")
tester = pd.read_csv("https://github.com/Obaida00/Housing-price-prediction/raw/main/test.csv")
PUPLIC_FEATURES = ["Id", "LotArea", "OverallQual", "OverallCond", "YearBuilt", "YearRemodAdd", "BsmtFullBath", "BsmtHalfBath", "FullBath", "HalfBath", "BedroomAbvGr", "KitchenAbvGr", "YrSold", "1stFlrSF", "2ndFlrSF", "SalePrice"]

trainer=trainer[PUPLIC_FEATURES].copy()
tester=tester[PUPLIC_FEATURES[:-1]].copy()

trainer.sort_values(by="YrSold", inplace=True)
trainer.reset_index(inplace=True, drop=True)
tester.sort_values(by="YrSold", inplace=True)
tester.reset_index(inplace=True, drop=True)

print("The Testing Dataframe sample...")
tester.head()
```

```python
train_features = trainer.drop(columns="SalePrice")
train_labels = trainer["SalePrice"]
```

```python
from catboost import CatBoostRegressor
import numpy as np

lol = CatBoostRegressor(verbose=False, allow_writing_files=False, random_state=0)
lol.fit(train_features, train_labels)
predicts = lol.predict(trainer)

predicts = np.round(predicts)
print(predicts)
```
